# 5. Recomendador basado en embeddings modernos

Implementad aquí vuestro modelo principal.

## Posibles estrategias

### Estrategia 1 — Ítem a ítem
- obtenéis embeddings de todos los ítems,
- calculáis similitud,
- y recomendáis los más cercanos.

### Estrategia 2 — Perfil de usuario
- agregáis embeddings de ítems consumidos / valorados positivamente,
- construís un embedding de usuario,
- y recomendáis por cercanía.

### Estrategia 3 — Query semántica
- representáis una necesidad del usuario como texto,
- la convertís a embedding,
- y recuperáis ítems cercanos.

### Estrategia 4 — Híbrido
- usáis embeddings para recuperar candidatos,
- y luego reordenáis con otra señal:
  popularidad, rating, collaborative score, filtros de negocio, etc.

## Qué debéis justificar

- modelo o servicio usado para obtener embeddings,
- coste / velocidad / facilidad de uso,
- dimensión de embeddings,
- función de similitud,
- si normalizáis o no,
- cómo hacéis retrieval,
- y por qué vuestra decisión tiene sentido para el dataset elegido.

```python
# Espacio de trabajo: generación de embeddings
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import normalize

# generate embedding
print("Loading embedding model...")
model = SentenceTransformer("all-MiniLM-L6-v2")   # lightweight, fast

print("Generating embeddings...")
embeddings = model.encode(dataset["Representation_text"], show_progress_bar=True)

# ensures that cosine similarity is computed correctly
embeddings = normalize(embeddings)

# compare similarities with cosine (the std)
print("Computing similarity matrix (embeddings)...")
cosine_sim_embeddings = cosine_similarity(embeddings, embeddings)

print("Embedding-based recommender ready")
```

```python
# Espacio de trabajo: recomendador con embeddings
def recommend_embeddings(game_title, dataset, sim_matrix, top_k=10):
    
    # Get index of the game
    idx = dataset[dataset["Game Title"] == game_title].index[0]
    
    # Get similarity scores
    scores = list(enumerate(sim_matrix[idx]))
    
    # Sort by similarity
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    
    # Remove the game itself
    scores = scores[1:top_k+1]
    
    # Get indices
    game_indices = [i[0] for i in scores]
    
    return dataset["Game Title"].iloc[game_indices]
```

We use the all-MiniLM-L6-v2 model to generate 384-dimensional embeddings for each item. Cosine similarity is used to compare items, and embeddings are normalized to improve stability. Recommendations are generated by computing pairwise similarities and retrieving the most similar items. This approach is efficient and suitable for the dataset, which contains rich item descriptions but no user interaction data.

- cómo definís el problema,
- cómo preparáis los datos,
- y cómo evaluáis.

## Consejo 3
Las mejoras pequeñas pueden ser reales… o ruido experimental.

Sed prudentes al interpretar resultados.

## Consejo 4
No confundáis:
- similitud semántica,
- recomendación útil,
- ranking,
- y predicción de preferencia.

Están relacionados, pero no son exactamente lo mismo.

## Consejo 5
Una muy buena práctica es incluir una pequeña tabla resumen final con:
- modelo,
- representación usada,
- métricas,
- ventajas,
- limitaciones.

# Extensiones opcionales

Si queréis subir nivel, podéis intentar alguna de estas extensiones:

- usar varias estrategias de construcción del string y compararlas,
- hacer un pequeño estudio de ablación,
- combinar embeddings con popularidad o collaborative filtering,
- añadir reranking,
- estudiar diversidad y cobertura además de precisión,
- crear una demo sencilla,
- analizar fallos,
- o proponer una versión para producción.

# Cierre

## Mensaje final

El objetivo de esta práctica no es demostrar que “lo moderno siempre gana”.

El objetivo es aprender a responder con datos, método y criterio a una pregunta muy real en sistemas de recomendación:

> **¿Cuándo merece la pena pasar de un enfoque clásico a uno basado en embeddings modernos?**

Buena suerte, y defended vuestras decisiones con claridad.
